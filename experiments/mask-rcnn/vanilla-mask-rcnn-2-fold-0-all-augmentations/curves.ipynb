```python
# load losses.txt to array
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('losses.txt')
# there are map50 and map75 columns. Calculate the average of them
df['map'] = (df['map50'] + df['map75']) / 2
# estimate map@60 by linear interpolation
df['map60'] = df['map50'] + (df['map75'] - df['map50']) / 0.25 * 0.1

# plot the map
plt.plot(df['map50'])
plt.ylim(0.6, 0.75)

# print argmax and max
print(np.argmax(df['map60']), np.max(df['map60']))
```
