```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load training_log.csv to a pandas dataframe
df = pd.read_csv('log.csv')
df
```

```python
# Plot loss and val_loss vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['epoch'], df['loss_train'], label='loss_train')
plt.plot(df['epoch'], df['loss_val'], label='loss_val')
plt.legend()

plt.ylim(0.2, 0.5)
```

```python
# print min loss_val and argmin loss_val
print(df['loss_val'].min())
print(df['loss_val'].argmin())
```
