```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load training_log.csv to a pandas dataframe
df = pd.read_csv('segformer.csv')
df
```

```python
# Plot loss and val_loss vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['epoch'], df['loss_train'], label='loss_train')
plt.plot(df['epoch'], df['loss_val'], label='loss_val')
plt.legend()

# plt.ylim(0.038, 0.04)
```

```python
# Plot loss and val_loss vs epoch
plt.figure(figsize=(10, 6))

plt.plot(df['epoch'], np.sqrt(df['loss_train'])/0.1, label='loss_train')
plt.plot(df['epoch'], np.sqrt(df['loss_val'])/0.1, label='loss_val')
plt.legend()

plt.ylim(0.5, 1.2)
```
