```python
import pandas as pd

df = pd.read_csv('runs/segment/train/results.csv')

# remove spaces in column names
df.columns = df.columns.str.replace(' ', '')

df
```

```python
# plot val/box_loss + 	val/seg_loss	+ val/cls_loss vs epoch
import matplotlib.pyplot as plt

plt.plot(df['epoch'], df['val/box_loss'] + df['val/seg_loss'] + df['val/cls_loss'])
```

```python
# plot metrics/mAP50(M)	vs epoch
import matplotlib.pyplot as plt

plt.plot(df['epoch'], df['metrics/mAP50(M)'])

# print the maximum mAP50(M) and the epoch it occurs
print(df['metrics/mAP50(M)'].max())
print(df['epoch'][df['metrics/mAP50(M)'].idxmax()])

plt.ylim(0.7)
```
